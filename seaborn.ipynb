```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
w_filepath = r'C:\Users\Lenovo\OneDrive\Desktop\seattle-weather.csv'
df = pd.read_csv(w_filepath)
df
```

```python
plt.figure(figsize = (15,6), dpi =100)

sns.lineplot(x=df['date'].head(10), y=df['temp_max'].head(10))
plt.show()
```

```python
plt.figure(figsize = (18,6), dpi = 100)

sns.barplot(x = df['date'].head(10), y = df['temp_max'].head(10))
plt.show()
```

```python
plt.figure(figsize = (20,6), dpi = 100)

sns.scatterplot(x = df['date'].head(20), y= df['temp_min'].head(20))
plt.show()
```

```python
sns.histplot(df['temp_max'], bins=20)
plt.show()
```

```python
sns.boxplot(y = df['temp_max'])
plt.show()
```

```python
sns.countplot(x=df['weather'])
plt.show()
```

```python
sns.heatmap(df[['temp_max','temp_min','wind']].corr(), annot = True,)
plt.show()
```

```python
plt.figure(figsize = (25,6), dpi = 200)

sns.lineplot(
    x=df['date'],
    y=df['temp_max'],
    hue=df['weather']
)
plt.show()
```

```python
plt.figure(figsize=(10,6),dpi = 200)
sns.stripplot(
    x = 'weather',
    y = 'temp_max',
    data = df,
    jitter = True
)
plt.title("Max Temperature by Weather Type")
plt.show()
```

```python
plt.figure(figsize = (10,6), dpi = 200)
sns.violinplot(
    x = 'weather',
    y = 'temp_max',
    data = df
)
plt.title("Distribution of Max Temperature by Weather")
plt.show()
```

```python
weather_counts = df['weather'].value_counts()

plt.figure(figsize = (10,6))
plt.pie(
    weather_counts,
    labels = weather_counts.index,
    autopct = '%1.1f%%',
    startangle = 90
)
plt.title("Weather Distribution")
plt.show()
```
